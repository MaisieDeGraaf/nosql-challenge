### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
penang_flavours = {"BusinessName": "Penang Flavours"}
degree_search = 0.01
latitude = establishments.find_one(penang_flavours)["geocode"]["latitude"]                           
longitude = establishments.find_one(penang_flavours)["geocode"]["longitude"] 

query = {
    "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search},
    "RatingValue": 5
} 

# Print the results
result = establishments.find(query).sort("scores.Hygiene", 1)
for establishment in result:
    print(f"Business: {establishment['BusinessName']}, Latitude: {establishment['geocode']['latitude']}, Longitude: {establishment['geocode']['longitude']}, Rating: {establishment['RatingValue']}, Hygiene: {establishment['scores']['Hygiene']}")
```

```python
# Convert result to Pandas DataFrame
PenangFlavours_competition = [x for x in establishments.find(query)]

PenangFlavours_competition_df = pd.DataFrame(PenangFlavours_competition)

print(f"Number of rows in DataFrame: {len(PenangFlavours_competition_df)}")
PenangFlavours_competition_df.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

match_query = {'$match': {'scores.Hygiene': 0}}

group_query = {'$group': {'_id': '$LocalAuthorityName', 'count': {'$sum': 1}}}

sort = {'$sort': {'count': -1}}


pipeline = [match_query, group_query, sort]

result = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
total_documents = sum(item['count'] for item in result)
print("Count of documents with a hygiene score of 0:", total_documents)

# Print the first 10 results
pprint(result[:10])
```

```python
# Convert the result to a Pandas DataFrame
Hygiene0_df = pd.DataFrame(list(result))

# Display the number of rows in the DataFrame
print(f"Number of rows in DataFrame: {len(Hygiene0_df)}")

# Display the first 10 rows of the DataFrame
Hygiene0_df.head(10)
```
